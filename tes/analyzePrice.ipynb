```python
def divide_conquer(df1, df2):
    if df1.shape[0] > 1:
        df1_half = df1.shape[0]//2
        df1 = divide_conquer(df1.iloc[:df1_half], df1.iloc[df1_half:])
    if df2.shape[0] > 1:
        df2_half = df2.shape[0]//2
        df2 = divide_conquer(df2.iloc[:df2_half], df2.iloc[df2_half:])
        
    if df1.shape[0] == 1:
        df1 = pd.DataFrame(calOrigin(df1))
    if df2.shape[0] == 1:
        df2 = pd.DataFrame(calOrigin(df2))

    return pd.concat([df1, df2], join='inner')


def calOrigin(sr):
    disType = sr['discount_type'][0]
    if disType[-1] == '%':
        rate = disType[-3:-1]
        if rate[-3:-1].isdecimal():
            origin = float(sr.iloc[0]['price']) * (100.0 / (100.0-float(rate)))
            
        else:
            origin = float(sr.iloc[0]['price']) * (100.0 / (100.0-float(rate[-1])))
    else:
        origin = np.nan

    return pd.concat([sr, pd.DataFrame([[origin]], index=sr.index, columns=['origin_price'])], axis=1)
```

#### 함수 사용 메뉴얼
- block     : 블록별로 확인한다
- row       : 열 별로 확인한다
- blocknrow : 각 블록의 열 별로 확인한다
- 1         : 할인이 적용된 데이터만 확인
- 2         : 원가로 구매한 데이터까지 확인
- 3         : 250000원 이하의 데이터만 확인
- 4         : 공연코드별로 확인

```python
def floorView_block(floor):
    df = seatsDF[seatsDF['floor'] == floor ]
    plt.figure(figsize=(18, 10))
    plt.scatter(df.origin_price, df.block, c='b', alpha=0.1, s=100)
    plt.xticks(fontsize=24)
    plt.yticks(fontsize=12)
```

```python
def floorView_block2(floor):
    df = seatsDF[seatsDF['floor'] == floor ]
    plt.figure(figsize=(18, 10))
    plt.scatter(df.origin_price, df.block, c='b', alpha=0.1, s=100)
    plt.scatter(noDis.price, noDis.block, c='g', alpha=0.1, s=100)
    plt.xticks(fontsize=24)
    plt.yticks(fontsize=12)
```

```python
def floorView_block3(floor):
    df = seatsDF[(seatsDF['floor'] == floor) & (seatsDF['origin_price'] <= 250000) ]
    ndf = noDis[noDis['price'] <= 250000]
    plt.figure(figsize=(18, 10))
    plt.scatter(df.origin_price, df.block, c='b', alpha=0.1, s=100)
    plt.scatter(ndf.price, ndf.block, c='g', alpha=0.1, s=100)
    plt.xticks(fontsize=24)
    plt.yticks(fontsize=12)
```

```python
def floorView_block4(floor, code):
    df = seatsDF[(seatsDF['floor'] == floor) & (seatsDF['performance_code'] == code) ]
    ndf = noDis[noDis['performance_code'] == code]
    date = df.iloc[0]['play_date']
    plt.figure(figsize=(18, 10))
    plt.scatter(df.origin_price, df.block, c='b', alpha=0.1, s=100)
    plt.scatter(ndf.price, ndf.block, c='g', alpha=0.1, s=100)
    plt.title(label=f'pcode : {code} / date : {date}')
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=12)
```
